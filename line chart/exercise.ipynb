```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
covid_data = pd.read_csv("owid-covid-data.csv", index_col = "date", parse_dates = True)
```

```python
sns.lineplot(data = covid_data)
```

```python
plt.figure(figsize= (16, 9))
plt.title("COVID-19 Cases")
sns.lineplot(data = covid_data)
```

```python
plt.figure(figsize= (16, 9))
plt.title("COVID-19 Cases")
plt.xlabel("Date")
plt.ylabel("Total Cases")
sns.lineplot(data = covid_data["total_cases"])
```
