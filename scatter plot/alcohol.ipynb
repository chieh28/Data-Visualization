```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
redwine_data = pd.read_csv("winequality-red.csv")
whitewine_data = pd.read_csv("winequality-white.csv")
```

```python
plt.title("Alcohol Quality")
ax1 = sns.scatterplot(x = redwine_data["fixed acidity"], y = redwine_data["alcohol"], size = 0.3, color='red')
ax2 = ax1.twinx()
ax2 = ax1.twiny()
ax2 = sns.scatterplot(x = whitewine_data["fixed acidity"], y = whitewine_data["alcohol"], size = 0.3, color='blue')
```

```python
plt.title("Alcohol Quality")
ax1 = sns.regplot(x = redwine_data["fixed acidity"], y = redwine_data["alcohol"], color='red')
ax2 = ax1.twinx()
ax2 = ax1.twiny()
ax2 = sns.regplot(x = whitewine_data["fixed acidity"], y = whitewine_data["alcohol"], color='blue')
```
