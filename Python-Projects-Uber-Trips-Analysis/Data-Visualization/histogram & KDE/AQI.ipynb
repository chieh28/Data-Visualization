```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
air_data = pd.read_csv("air quality.csv", index_col = "aqi")
sns.histplot(air_data.index)
```

```python
sns.kdeplot(air_data.index, shade = True)
```

```python
sns.jointplot(x = air_data.index, y = air_data["pm2.5_avg"], kind = "kde", shade = True)
```

```python
plt.figure(figsize = (16, 9))
sns.histplot(data = air_data, x = "pm2.5_avg", hue = "county")
```

```python
plt.figure(figsize = (16, 9))
sns.kdeplot(data = air_data, x = "pm2.5_avg", hue = "county")
```
