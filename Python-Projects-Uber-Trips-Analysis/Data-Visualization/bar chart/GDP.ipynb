```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
GDP_data = pd.read_csv("gdp.csv", index_col = "2021")
rank_GDP_data = GDP_data.sort_values(by = "2021", ascending = False).head(10)
```

```python
plt.figure(figsize = (10, 6))
plt.title("top 10 countriesâ€™ GDP in 2021")
sns.barplot(x = rank_GDP_data.index, y = rank_GDP_data["Country Name"])
plt.xlabel("GDP(current US$)")
plt.ylabel("Country")
```
